**10. Get a subset of Cleveland Caveliers (CLE), the Runner-up team in the season. Who is the highest pay of CLE? Who is the lowest? How many times the highest-pay player than more than the lowest? (3 marks)**

```python
cle = nba[nba['team'] == 'CLE']
cle.sort_values('salary')
```

```python
cle[cle['salary'] == cle['salary'].max()]
```

```python
cle[cle['salary'] == cle['salary'].min()]
```

```python
cle['salary'].max()/cle['salary'].min()
```

**Answer: The highest pay player in CLE is Lebron James, while the lowest pay player is Ante Zizic. The salary of Lebron James is 20 times of Ante Zizic's salary.**
